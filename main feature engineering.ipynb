```python
import pandas as pd
# import re
# from sklearn.preprocessing import MultiLabelBinarizer
```

```python
df_train = pd.read_csv("data/train.csv", delimiter=",")
df_test = pd.read_csv("data/test.csv", delimiter=",")
```

```python
df_train.isnull().sum()
```

```python
df_test.isnull().sum()
```

```python
df_train['Embarked'] = df_train['Embarked'].fillna(df_train['Embarked'].mode()[0],)
df_train['Age'] = df_train['Age'].fillna(df_train['Age'].median())
df_test['Embarked'] = df_test['Embarked'].fillna(df_train['Embarked'].mode()[0],)
df_test['Age'] = df_test['Age'].fillna(df_train['Age'].median())
df_test['Fare'] = df_test['Fare'].fillna(df_train['Fare'].median())
```

```python
x_train_wName = df_train[['Pclass','Name','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked']]
X = df_train[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked','Ticket']]

y = df_train['Survived']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
X['Has_Ticket_Prefix'] = X['Ticket'].str.match(r"[^\d]+").astype(int)
X['Ticket_Number'] = X['Ticket'].str.replace('^[A-Za-z/\.]+', '', regex=True).str.strip()
X['Starting_Digit'] = X['Ticket_Number'].str.extract('^(\d)', expand=False)
X['Ticket_Number_Length'] = X['Ticket'].str.replace(r'[^\d]+', '', regex=True).apply(len)
X.drop(columns =['Ticket', 'Ticket_Number'], inplace =True)  # Drop original column (optional)
```

```python
X['Starting_Digit'] = X['Starting_Digit'].fillna(X['Starting_Digit'].mode()[0],)
```

```python
X.isnull().sum()
```

```python
x_train_encoded = pd.get_dummies(X, columns=['Pclass','Sex', 'Embarked'], drop_first=False).astype(int)
# X = X.drop(['Pclass','Sex', 'Embarked'], axis=1)
# X = pd.concat([X, x_train_encoded], axis=1)
```

```python
print("X_train_encoded:\n", x_train_encoded)
```

```python
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, cross_validate
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
log_reg = LogisticRegression(random_state=42, max_iter=10000, )
skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
scores_stratified = cross_val_score(log_reg, x_train_encoded, y, cv=skf, scoring='f1') # 'f1' is often a good choice for imbalanced data
print("Stratified Cross-validation scores:", scores_stratified)

log_reg.fit(x_train_encoded, y)
```

```python
x_score = df_test[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked', 'Ticket']]

x_score['Has_Ticket_Prefix'] = x_score['Ticket'].str.match(r"[^\d]+").astype(int)
x_score['Ticket_Number'] = x_score['Ticket'].str.replace('^[A-Za-z/\.]+', '', regex=True).str.strip()
x_score['Starting_Digit'] = x_score['Ticket_Number'].str.extract('^(\d)', expand=False)
x_score['Ticket_Number_Length'] = x_score['Ticket'].str.replace(r'[^\d]+', '', regex=True).apply(len)
x_score.drop(columns =['Ticket', 'Ticket_Number'], inplace =True)  # Drop original column (optional)

x_score_encoded = pd.get_dummies(x_score, columns=['Pclass','Sex', 'Embarked'], drop_first=True).astype(int)

# Align columns: add missing columns in x_score with NaN
x_score_encoded = x_score_encoded.reindex(columns=x_train_encoded.columns, fill_value=0)

# x_score = x_score.drop(['Pclass','Sex', 'Embarked'], axis=1)
# x_score = pd.concat([x_score, x_score_encoded], axis=1)
```

```python
y_pred = log_reg.predict(x_score_encoded)
```

```python
y_pred
```

```python
submission = pd.Series(y_pred, index=df_test['PassengerId'], name='Survived').reset_index()
```

```python
submission.to_csv('pred_result/submission.csv', index=False)
```
